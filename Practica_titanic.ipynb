Vemos que las celdas vacias, estan en la columna Age (Edad), Cabin (Cabina) y Embarked(Puerto de embarque).
* He decidido llenar los dato de Age con el prommedio de edades en total
* Eliminar la columna Cabin, por la mayoria de falta de datos y la poca relacion para determinar su valor.
* La columna Embarked llenar los valores vacios con 0, dado que no hay forma de encontrar una forma de rescatar estos valores. No se eliminan porque de alguna forma se sabe que estos pasajeros existieron por el boleto comprado y no afecta resolver esta practica.

### 2. ¿Cuál es el porcentaje de personas que sobrevivió?
Tomaremos Survived (1) or died (0).
El .mean() nos dará el la suma de todos los valores dividido entre el total, si a eso lo multiplicamos por 100, nos dará el porcentaje de sobrevivientes.

```python
df["Survived"].value_counts()
```

```python
print("El porcentaje de sobrevivientes fue de:",(df["Survived"].mean()*100).round(2),"%")
```

### 4. ¿Cuál es el porcentaje de gente que sobrevivió que pagó un pasaje de menos de 10?
Primero generaremos un nuevo dataframe solo con los sobrevivientes. Sumamos los valores de la columna Fare cuando son menores de 10, multiplicamos por 100 y dividimos en el total

```python
df_surv = df[df["Survived"] != 0]
df_surv.info()
```

```python
df_surv.head()
```

```python
menor10 = 100*((df_surv["Fare"]<10).sum())/(df_surv["Survived"].sum())
print("El porcentaje de pasajeros que sobrevivio y pago un pasaje menor a 10 fue de:",menor10.round(2),"%")
```
